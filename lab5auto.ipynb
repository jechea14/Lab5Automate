```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('C:\\Users\\Jeanie\\OneDrive\\python\\python_csv\\146lab5auto\\data1.csv')
```

```python
data
```

```python
d = {'Avg Energy': [data['SW Energy'].mean(), data['HW Energy'].mean(), data['DMA Energy'].mean(), data['DMA-LPM Energy'].mean()]}

d1 = {'Avg Mean Current': [data['SW Mean Current'].mean(), data['HW Mean Current'].mean(), data['DMA Mean Current'].mean(), data['DMA-LPM Mean Current'].mean()]}

df = pd.DataFrame(d, ["SW", "HW", "DMA", "DMA-LPM"])
df1 = pd.DataFrame(d1, ["SW", "HW", "DMA", "DMA-LPM"])
```

```python
df
```

```python
df1
```

```python
# ax = df.plot.bar(subplots=True, rot=0, figsize=(9,9))
ax = df.plot.bar(rot=0)
ax.set_xlabel("Method")
ax.set_ylabel("Energy (mJ)")
ax.set_title("Energy")

for p in ax.patches:
    ax.annotate("%.4f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.savefig('Energy.png')
```

```python
ax = df1.plot.bar(rot=0)
ax.set_xlabel("Method")
ax.set_ylabel("Mean Current (mA)")
ax.set_title("Mean Current")

for p in ax.patches:
    ax.annotate("%.4f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.savefig('Mean_Current.png')
```

```python
with pd.ExcelWriter('C:\\Users\\Jeanie\\OneDrive\\python\\python_csv\\146lab5auto\\data2.xlsx') as writer:
    data.to_excel(writer, sheet_name='Data')
    df.to_excel(writer, sheet_name='DataCalculations', startrow=0, startcol=0)
    df1.to_excel(writer, sheet_name='DataCalculations', startrow=df.shape[0] + 5, startcol=0)
```
